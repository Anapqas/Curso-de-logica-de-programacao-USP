```python
#2week6ex2
#Implemente a função encontra_impares(lista), 
#que recebe como parâmetro uma lista de números inteiros 
#e devolve uma outra lista apenas com os números ímpares da lista dada.

def encontra_impares(lista, resposta = None):
    if resposta == None:
        resposta = []
    n = len(lista)
    if n <= 1 and lista[0]%2 != 0  :
        resposta.append(lista[0])
        return resposta
    elif n <= 1 and lista[0]%2 == 0  :
        return resposta
    elif lista[0]%2 != 0:
        resposta.append(lista[0])
        encontra_impares(lista[1:], resposta) 
        return resposta
    elif lista[0]%2 == 0:
        return encontra_impares(lista[1:], resposta)
```
